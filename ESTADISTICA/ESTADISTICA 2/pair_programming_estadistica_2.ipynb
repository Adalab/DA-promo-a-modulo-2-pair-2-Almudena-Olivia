EJERCICIO 1:
Extraer las siguentes estadÃ­siticas del conjunto de datos para varias de las varibles disponibles:

- Utilizando mÃ©todos de pandas: media, mediana, moda, mÃ¡ximo y mÃ­nimo
    

```python
age = df['age']
```

```python
edades_bien = df.copy()
```

```python
edades_bien.dropna(subset = ["age"], inplace=True)
```

```python
age_bien = edades_bien['age']
```

```python
media = np.mean(age_bien)
print('La media de edad es de:', media)
```

```python
mediana = np.median(age_bien)
print('La mediana es:', mediana)
```

```python
moda, frecuencia = stats.mode(age_bien)
print('La moda de edad es:', moda,'y la frecuencia es:', frecuencia)
```

```python
minimo = np.min(age_bien)
print('La edad mÃ­nima es:', minimo)
```

```python
maximo = np.max(age_bien)
print('La edad mÃ¡xima es:', maximo)
```

```python
q3, q1 = np.percentile(age_bien, [75, 25])
print('El percentil 25 es:', q1)
print('El percentil 75 es:', q3)
```

```python
riq = q3 - q1
print("El rango intercuartÃ­lico es:", riq)
```

- Tabla de frecuencias
    - Variables cualitativas
    - Variables cuantitativas
    - Frecuencia acumulada
    - RepresentaciÃ³n de al menos una tabla de frecuencias

    - variable cualitativa

```python
edades_categoria = edades_bien.groupby("category") 
print(edades_categoria)

frecuencias_categoria = edades_categoria.agg(frecuencia=("category", "count")) 
print(frecuencias_categoria)
```

    - variable cuantitativa

```python
edad_cuant = edades_bien.groupby("age")
frec_edad =  edad_cuant.agg(frecuencia=("age", "count"))
print(frec_edad)
```

    - frecuencia acumulada

```python
bins = np.arange(19, 100+10, 10)

edades_bien["age_group"] = pd.cut(edades_bien["age"], bins=bins)

edades_cut = edades_bien.groupby("age_group")
edades_cut_bins =  edades_cut.agg(frecuencia=("age", "count"))
edades_cut_bins["frecuencia_acumulada_edad"] = edades_cut_bins["frecuencia"].cumsum()
edades_cut_bins
```

    - RepresentaciÃ³n de al menos una tabla de frecuencias

```python
plt.rcdefaults()

df_categoria = edades_bien.groupby("category") 
frecuencias_categoria = df_categoria.agg(frecuencia=("category", "count")) 
frecuencias_categoria.plot.bar()
```

```python
plt.rcdefaults()

df_genero = edades_bien.groupby("gender") 
frecuencias_genero = df_genero.agg(frecuencia=("gender", "count")) 
frecuencias_genero.plot.bar()
```

```python
plt.rcdefaults()
edades_bien["age_group"] = pd.cut(edades_bien["age"], bins=20)
df_edad = edades_bien.groupby("age") 
frecuencias_edad = df_edad.agg(frecuencia=("age_group", "count")) 
frecuencias_edad.plot.bar(figsize=(20, 6))

```
