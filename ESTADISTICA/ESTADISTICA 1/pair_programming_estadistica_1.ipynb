EJERCICIO 2: Extraer las siguentes estadÃ­siticas del conjunto de datos para varaias de las varibles disponibles:

```python
finalWorth = df['finalWorth']
```

```python
media = np.mean(finalWorth)
print('La media de patrimonio es de:', media)
```

```python
desviacion_media = np.abs(finalWorth-media)
print('La desviacion media de patrimonion es de:', '\n', desviacion_media)
```

```python
varianza = np.var(finalWorth)
print('La varianza de los datos del patrimonio es de:', varianza)
```

```python
desviacion_tipica = np.std(finalWorth)
print('La desviaciÃ³n tÃ­pica es de:', desviacion_tipica)
```

```python
moda, frecuencia = stats.mode(finalWorth)
print('La moda del patrimonio es:', moda,'y la frecuencia es:', frecuencia)
```

```python
mediana = np.median(finalWorth)
print('La mediana es:', mediana)
```

```python
finalWorth.shape
```

```python
w = np.ones(2668, dtype=np.complex64)
avg = np.average(finalWorth, weights=w)
print('La media ponderada es:', avg)
```

```python
coeficiente_variacion = desviacion_tipica/ media
print('El coeficiente de variaciÃ³n es:', coeficiente_variacion)
print('Los datos no son robustos porque el coeficiente de variaciÃ³n es muy alto')
```
