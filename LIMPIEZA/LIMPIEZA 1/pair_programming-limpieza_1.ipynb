EJERCICIO 3: Como hemos visto, algunas columnas no tienen el tipo de datos que deberían. Cambiad el tipo de dato para la columna de year.

```python
attacks['year'].dtypes
```

```python
attacks['year'] = attacks['year'].astype('int64', errors='ignore')
attacks['year'].dtypes
```

EJERCICIO 4: En la columna de country poner todos los valores en minúsculas.💡 Pista Tendréis que usar una función o una lambda 💡

```python
def lower_case (x):
    try:
        return x.lower()
    except:
        return np.NaN           #En la columna 'country' hay NaN y esos son tipo float. 
```

```python
attacks["country"] = attacks["country"].apply(lower_case)
attacks.head()
```
