 - Explorad la columna de country y reportad los resultados solo para aquellas categorías que incluyan el 70% de los datos.

```python
df.stb.freq(['country'], thresh=70)
```

    - Podemos observar que son muy pocos los países que tengan más del 70% de datos completos, estando USA y Australia a la cabeza.

 - Explorad la columna country en relación al sexo, filtrando de nuevo por aquellas categorías que incluyan el 70% de los datos.

```python
df.stb.freq(["country", 'sex_'], thresh=70)
```

    - Esta tabla de frecuencia nos muestra que la mayoría de ataques son hacia hombres. 

 - Haced los mismo para la columna activity en relación a la columna fatal. En este caso filtrad y mostrad los datos que incluyan el 60% de los datos.

```python
df.stb.freq(["activity", 'fatal_(y/n)'], thresh=60)
```

    - Haciendo un análisis de frecuencia entre la columna activity y fatal_(y/n), la mayoría de los ataques no han sido mortales y se han producido mientras nadaban, surfeaban y pescaban. Cuando sí ha sido mortal el ataque, este se produjo mientras la víctima nadaba o se bañaba en el mar.
