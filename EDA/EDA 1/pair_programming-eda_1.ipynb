 - ¿Cuál es el porcentaje de valores nulos en cada columna? Presenta los resultados en un DataFrame, donde las columnas sean el nombre de las variables y el porcentaje de nulos.

```python
nulos = df.isnull().sum()
porcentaje_isnull = (nulos/ len(df)*100)
porcentaje_isnull = np.round(porcentaje_isnull, decimals = 2)
nuls = pd.DataFrame(porcentaje_isnull)

nuls.rename(columns = {0:'porcentaje nulos'}, inplace = True)
nuls
```
